```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
# reflect an existing database into a new model

## Declare a Base class to reflect the database tables
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect = True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Station = Base.classes.station
Measurement = Base.classes.measurement
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
session.query(Measurement.date).order_by(Measurement.date.desc()).first()
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
```

```python
# Calculate the date one year from the last date in data set.
OneYrAgo = dt.date(2017,8,23) - dt.timedelta(days=365)
print(OneYrAgo)
```

```python
# Perform a query to retrieve the data and precipitation scores
sel = [Measurement.date,Measurement.prcp]
session.query(*sel).\
    filter(Measurement.date >= OneYrAgo).\
    order_by(Measurement.date.desc()).all()
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
# Sort the dataframe by date
Last12Months = session.query(*sel).filter(Measurement.date >= OneYrAgo).order_by(Measurement.date.desc()).all()
precip_df = pd.DataFrame(Last12Months, columns=["Date","Precipitation"])
precip_df.head()
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
x_axis = precip_df.Date.tolist()
y_axis = precip_df['Precipitation'].tolist()

plt.figure(figsize=(15,7))
plt.plot(x_axis,y_axis,color = 'r')
plt.title(f"Precipitation from {OneYrAgo} to {dt.date(2017,8,23)}")
plt.xlabel("Date")
plt.ylabel("Precipitation")

plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
precip_df.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
session.query(Station.station).count()

## Station table contains a list of all the stations.  This is the number of unique station id.
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
sel = [Measurement.station,
      func.count(Measurement.station)]

Station_Activity = session.query(*sel).\
group_by(Measurement.station).\
order_by(func.count(Measurement.station).desc()).all()

Station_Activity
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
sel = [Measurement.station, Measurement.date, Measurement.tobs]
Station_USC00519281 = session.query(*sel).filter(Measurement.station == "USC00519281").order_by(Measurement.date.desc()).all()
Station_USC00519281_df = pd.DataFrame(Station_USC00519281, columns=["Station","Date","Temperature"])

St_USC00519281_Min = Station_USC00519281_df.Temperature.min()
St_USC00519281_Max = Station_USC00519281_df.Temperature.max()
St_USC00519281_Ave = Station_USC00519281_df.Temperature.mean()

print(f"For station USC00519281, the lowest recorded temperature is {St_USC00519281_Min},\nthe highest recorded temperature is {St_USC00519281_Max}, and the average temperature is {St_USC00519281_Ave}")
```

```python
## Using the func.min, func.max, and func.avg functions to obtain the abovementioned task:
sel = [func.min(Measurement.tobs),func.max(Measurement.tobs),func.avg(Measurement.tobs)]
St_USC00519281_MinMaxAve = session.query(*sel).\
    group_by(Measurement.station).\
    order_by(func.count(Measurement.station).desc()).first()
St_USC00519281_MinMaxAve
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
```

```python
## Determine the latest date recorded for station USC00519281
Station_USC00519281_df.head(1)
```

```python
## Calculate date 12 months ago
OneYrAgo_USC00519281 = dt.date(2017,8,18) - dt.timedelta(days=365)
print(OneYrAgo_USC00519281)
```

```python
## Create Dataframe for station USC00519281 from 2016-08-18 to 2017-08-18
sel = [Measurement.date, Measurement.tobs]
Station_USC00519281_12mth = session.query(*sel).filter(Measurement.station == "USC00519281").filter(Measurement.date >= OneYrAgo_USC00519281).order_by(Measurement.date.desc()).all()
Station_USC00519281_12mth_df = pd.DataFrame(Station_USC00519281_12mth, columns=["Date","Temperature"])
Station_USC00519281_12mth_df
```

```python
## Plot histogram of station USC00519281's temperature records from 2016-08-18 to 2017-08-18


Station_USC00519281_12mth_df.plot.hist(alpha=0.5)

plt.title("Histogram of Temperature Recorded on Station USC00519281 from 2016-08-18 to 2017-08-18")
plt.xlabel("Temperature")
```
